```python
import os
from dotenv import load_dotenv
from stt import listen
from ttsmms import TTS
from IPython.display import Audio, display
import time
import threading
import queue

import google.generativeai as genai


load_dotenv()
GOOGLE_API_KEY = os.environ.get('GOOGLE_API_KEY')
genai.configure(api_key=GOOGLE_API_KEY)

model = genai.GenerativeModel('gemini-pro',)


response = model.generate_content(listen())
print(response.text)

# Load the model
tts = TTS("data/awa")

def synthesize(text, q):
    # Synthesize speech
    wav = tts.synthesis(text)
    q.put(wav)

texts = (response.text).split('।')


# Create a queue to communicate between threads
q = queue.Queue()

# Start a thread to play audio
def play_audio():
    while True:
        wav = q.get()
        if wav is None:
            break
        display(Audio(wav["x"], rate=wav["sampling_rate"], autoplay=True))
        # Wait for the audio to finish playing
        time.sleep(len(wav["x"]) / wav["sampling_rate"] + 1)  # Adding a buffer time

audio_thread = threading.Thread(target=play_audio)
audio_thread.start()

# Synthesize and play each text
for text in texts:
    synthesize(text, q)

# Signal the end of text stream
q.put(None)

```
