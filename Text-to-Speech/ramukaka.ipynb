```python
from ttsmms import TTS
tts=TTS("data/awa")
```

```python
wav=tts.synthesis('''भारत के ऐसे कई प्रसिद्ध उपन्यासकार हैं जिनकी रचना ने अमरत्व को प्राप्त किया है और जिन्होंने अपनी लेखनी से समाज के मुद्दों को सरकार से लेकर दुनियां तक को बताया है। एक ऐसे ही रचनाकार हैं, प्रेमचंद ! जिन्होंने अपनी लेखनी से गरीब-शोषित लोगों के दर्द को दर्शाया और उनके प्रसिद्ध उपन्यास विश्व में विख्यात हुए है।''')
```

```python
from IPython.display import Audio
Audio(wav["x"], rate=wav["sampling_rate"])
```
